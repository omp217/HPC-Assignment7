```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot(file_name, machine, problem_name):
    df = pd.read_csv(file_name)
    cores = df['NUMBER_CORES'].unique()
    df['PROB_SIZE'] = np.log2(df['PROB_SIZE'])

    for core in cores:
        plt.plot(df[df['NUMBER_CORES'] == core]['PROB_SIZE'], df[df['NUMBER_CORES'] == core]['ALG_TIME'], label = 'No. of cores = ' + str(core))

    plt.xlabel('Problem Size')
    plt.ylabel('Algorithm Time')
    plt.title('Algorithm Time vs Problem Size\n' + problem_name + '\n' + machine)
    plt.grid()
    plt.legend()
    plt.show()
```

# LAB

```python
machine = 'LAB 207'
problem_name = 'PI_RANDOM'
```

```python
file_name = 'data_lab_1.csv'
plot(file_name, machine, problem_name)
```

```python
file_name = 'data_lab_2.csv'
plot(file_name, machine, problem_name)
```
